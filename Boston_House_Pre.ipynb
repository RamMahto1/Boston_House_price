```python
## boston project - regressor problem
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.data
```

```python
print(boston.target)
```

```python
print(boston.DESCR)
```

```python
## CREATE THE DATA FRAME
dataset=pd.DataFrame(boston.data,columns=boston.feature_names)
dataset.head()
```

```python
dataset['Price']=boston.target
```

```python
## Spit the into dependent and independent
dataset.head()
```

```python
X = dataset.drop('Price',axis=1)
y= dataset['Price']
print('shape of the X',X.shape)
print('shape of the y',y.shape)
```

```python
## split the data into train and test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=41)
print('shape of the X_train',X_train.shape)
print('shape of the X_test',X_test.shape)
print('shape of the y_train',y_train.shape)
print('shape of the y_test',y_test.shape)
```

```python
## feature scalling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(X_train)
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
## train the model
from sklearn.linear_model import LinearRegression
reg_model = LinearRegression()
reg_model.fit(X_train,y_train)
```

```python
## check the coffience and intercept
print(reg_model.coef_)
```

```python
print(reg_model.intercept_)
```

```python
## find the best param
print(reg_model.get_params)
```

```python
## prediction
y_pred=reg_model.predict(X_test)
print(y_pred)
```

```python
## plot y_test and y_pred
plt.scatter(y_test,y_pred)
```

```python
sns.regplot(y_test,y_pred)
```

```python
## find the residual 
residual = y_test - y_pred
print(residual)
```

```python
sns.distplot(residual)
```

```python
## ploting the residual
sns.displot(residual,kind='kde')
```

```python
## scatter plot with respect to y_pred and residual
sns.scatterplot(y_pred,residual)
```

```python
## confusion matrix
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_absolute_error(y_test,y_pred)))
```
